## Ordinal Encoding on education

```python
dfmovie
```

```python
dfmovie.columns
```

```python
dfmovie['education'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord=OrdinalEncoder()
```

```python
ord
```

```python
ord.fit(dfmovie[['education']])
```

```python
ordary=ord.transform(dfmovie[['education']])
```

```python
ordary
```

```python
dfedu=pd.DataFrame(ordary,columns=['Degree'],dtype='int')
```

```python
dfedu
```

```python
ord.categories_ # By default PG is considered as 0, School as 1 and UG as 2
```

```python
dfmovie=pd.concat([dfmovie,dfedu],axis=1)
```

```python
dfmovie
```

```python
dfmovie=dfmovie.drop('education',axis=1)
```

```python
dfmovie
```

## OneHotEncoding on gender

```python
ohe=OneHotEncoder(drop='first')
```

```python
ohe
```

```python
ohe.fit(dfmovie[['gender']])
```

```python
oheary=ohe.transform(dfmovie[['gender']]).toarray()
```

```python
oheary
```
