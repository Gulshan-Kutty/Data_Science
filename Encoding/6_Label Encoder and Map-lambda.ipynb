## Label Encoding on review

```python
dfmovie.columns
```

```python
dfmovie['review'].unique()
```

```python
dfmovie['review'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
label.fit(dfmovie['review'])
```

```python
rev_ary=label.transform(dfmovie['review'])
```

```python
rev_ary
```

```python
dfrev=pd.DataFrame(rev_ary,columns=["Review"])
```

```python
dfrev
```

```python
dfmovie.head()
```

```python
dfrev.head()
```

```python
dfmovie=pd.concat([dfmovie,dfrev],axis=1).drop('review',axis=1)
```

```python
dfmovie
```

## Ordinal Encoding on education

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord=OrdinalEncoder(categories=[['School','UG','PG']])
```

```python
ord.fit(dfmovie[['education']])
```

```python
eduary=ord.transform(dfmovie[['education']])
```

```python
dfedu=pd.DataFrame(eduary,columns=['Education'],dtype='int')
```

```python
dfmovie=pd.concat([dfmovie,dfedu],axis=1).drop('education',axis=1)
```

```python
dfmovie
```
