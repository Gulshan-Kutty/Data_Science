## OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
dftown1=dftown=pd.read_csv(r'C:\Users\User\Downloads\Outlier Detection Dataset\Onehotencoding dataset\township.csv')
```

```python
dftown1
```

```python
ohe1=OneHotEncoder(drop='first')
```

```python
ohe1.fit(dftown1[['town']])
```

```python
oheary=ohe1.transform(dftown1[['town']]).toarray()
```

```python
oheary
```

```python
ohe1.categories_
```

```python
collabel=['robinsville', 'west windsor']
```

```python
dfencode=pd.DataFrame(oheary,columns=collabel,dtype='int')
```

```python
dfencode
```

```python
dftown1=pd.concat([dftown,dfencode],axis=1).drop('town',axis=1)
```
