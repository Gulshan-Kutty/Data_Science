```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
wine
```

```python
wine.keys()
```

```python
wine['data']
```

```python
pd.DataFrame(wine['data'])
```

```python
dfwine=pd.DataFrame(wine['data'],columns=['alcohol',
  'malic_acid',
  'ash',
  'alcalinity_of_ash',
  'magnesium',
  'total_phenols',
  'flavanoids',
  'nonflavanoid_phenols',
  'proanthocyanins',
  'color_intensity',
  'hue',
  'od280/od315_of_diluted_wines',
  'proline'])
```

```python
dfwine
```

```python
wine['target']
```

```python
dfwine['Target']=wine['target']
```

```python
dfwine
```

```python
wine['target_names']
```

```python
dfwine
```

```python
X=dfwine.drop('Target',axis=1)
```

```python
X
```

```python
y=dfwine['Target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.20)
```

```python
Xtrain.shape,Xtest.shape,X.shape
```

```python
ytrain.shape,ytest.shape,y.shape
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada=AdaBoostClassifier()
```

```python
ada
```

```python
ada.fit(Xtrain,ytrain)
```

```python
ada.score(Xtest,ytest)
```

## SVM & Logistic with Ada Boost Base Estimators

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.svm import SVC
```

```python
ajsvm=SVC()
```

```python
ada_svm=AdaBoostClassifier(base_estimator=ajsvm,n_estimators=50,algorithm='SAMME')
```

```python
ada_svm
```

```python
ada_svm.fit(Xtrain,ytrain)
```

```python
ada_svm.score(Xtest,ytest)
```

```python
ada_svm.score(Xtrain,ytrain) # its a underfitting case
```
