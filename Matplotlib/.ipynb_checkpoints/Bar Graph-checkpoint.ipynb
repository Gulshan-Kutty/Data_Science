```python
X = ["APPLE","ORANGE", "MANGO", "BANANA"]
y = [20,45,30,60]
```

```python
days =  np.arange(1,31)
temp = np.random.randint(20, 35,(30))
humid = np.random.randint(30, 45,(30))
```

```python
days
```

```python
temp
```

```python
humid
```

```python
plt.bar(days, temp,color = "b")
plt.xlabel('DAYS')
plt.ylabel("TEMPERATURE")
plt.title("HYDERABAD TEMPERATURE")
plt.show()
```

```python
plt.bar(days,humid, color = "y")
plt.xlabel('DAYS')
plt.ylabel("HUMIDITY")
plt.title("HYDERABAD TEMPERATURE")
plt.show()
```

```python
plt.bar(days,humid, color = "y")
plt.bar(days, temp,color = "b")
plt.xlabel('DAYS')
plt.ylabel("HUMIDITY & TEMPERATURE")
plt.title("HYDERABAD TEMPERATURE")
plt.show()
```

```python
days =  np.arange(1,31)
dtemp = np.random.randint(20, 35,(30))
htemp = np.random.randint(30, 40,(30))
```

```python
plt.bar(days, htemp, color = "r", label = "Hyd")
plt.bar(days, dtemp,color ="b", label = "Del")
plt.legend(loc=1)
```

```python
plt.bar(days, htemp, color = "r", label = "Hyd")
plt.barh(days, dtemp,color ="b", label = "Del")
plt.legend(loc=1)
```
