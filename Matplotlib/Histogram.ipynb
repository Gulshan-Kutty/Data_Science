```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt
```

```python
pystd = np.random.randint(20, 50,(80))
```

```python
aistd = np.random.randint(20,50,(80))
```

```python
plt.hist(pystd, width = 1.2)
plt.show()
```

```python
plt.hist(pystd,bins=[20,25,30,35], width = 4.5)
plt.show()
```

```python
plt.hist(pystd,bins=[20,25,30,35], rwidth = 0.95)
plt.show()
```

```python
plt.hist(pystd,bins=[20,25,30,35], rwidth = 0.95,color = "y")
plt.show()
```

```python
plt.hist(pystd,bins= 10,color = "y", rwidth=0.8)
plt.show()
```

```python
plt.hist(pystd,bins= 10,color = "b", rwidth=0.8, histtype="step")
plt.show()
```

```python
plt.hist(pystd,bins= 10,color = "b", rwidth=0.8, histtype="stepfilled")
plt.show()
```

```python
plt.hist(pystd,bins= 10,color = "b", rwidth=0.8, histtype="bar")
plt.show()
```

```python
plt.hist(pystd,bins= 10,color = "y", rwidth=0.8, orientation= "horizontal")
plt.show()
```

```python
plt.hist(pystd,bins= 10,color = "y", rwidth=0.8, orientation= "horizontal", label = "Python")
plt.legend(loc = 1)
plt.show()
```

```python
plt.hist(aistd,bins= 10,color = "y", rwidth=0.8, orientation= "horizontal", 
         label = "AI")
plt.hist(pystd,bins= 10,color = "b", rwidth=0.8)

plt.legend(loc = 1)
plt.show()
```

```python
plt.style.use("ggplot")
plt.hist([pystd, aistd], bins= 5, color = ["r","b"], rwidth=0.8, 
         label = ['Python', "AI"])
plt.legend(loc = 1)
plt.show()
```
