## Vectorization(Encoding)

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
```

```python
cvec=CountVectorizer()
```

```python
cvec.fit(dfimdb['rev_clean'])
```

```python
cvec.get_feature_names_out()
```

```python
cvec.vocabulary_
```

```python
rev_cvec=cvec.transform(dfimdb['rev_clean'])
```

```python
rev_cvec
```

```python
rev_cvec.toarray()
```

```python
rev_cvec
```

```python
tftrans=TfidfTransformer()
```

```python
tftrans
```

```python
tftrans.fit(rev_cvec) # output of count vectorizer should be input to this
```

```python
rev_tf=tftrans.transform(rev_cvec)
```

## Model Building

```python
X=rev_tf.copy()
```

```python
y=dfimdb['Sentiment']
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb=MultinomialNB()
```

```python
mnb
```

```python
mnb.fit(Xtrain,ytrain)
```

```python
mnb.score(Xtest,ytest)
```

## Prediction

```python
sent=['Love this product, its quality is very good']
sent_cvec=cvec.transform(sent)
sent_tf=tftrans.transform(sent_cvec)
output=mnb.predict(sent_tf)
output
```
