```python
dfsales=pd.read_csv(r'C:\Users\User\Downloads\DATASETS\Crosstab\Crosstab_Sales.csv')
```

```python
dfsales
```

```python
dfsales=dfsales.drop('Date',axis=1)
```

```python
dfsales
```

```python
dfsales['Sales']=dfsales["Sales"].replace('[$,]','',regex=True)
```

```python
dfsales
```

```python
dfsales.dtypes
```

```python
dfsales['Sales']=dfsales["Sales"].astype('float')
```

```python
dfsales.dtypes
```

```python
dfsales.head()
```

```python
pd.crosstab(index = dfsales["Region"], columns = dfsales["Type"])
```

```python
pd.crosstab(dfsales['Region'], dfsales["Type"])
```

```python
pd.crosstab(dfsales['Region'], dfsales["Type"], values=dfsales["Units"], aggfunc="mean")
```

```python
pd.crosstab(dfsales['Region'], dfsales["Type"], values=dfsales["Units"], 
            aggfunc="mean",margins=True)
```

```python
pd.crosstab(index= dfsales["Type"], columns =  dfsales["Region"],
            values=dfsales["Sales"], aggfunc="sum")
```

```python
pd.crosstab(index= dfsales["Type"], columns =  dfsales["Region"],
            values=dfsales["Sales"], aggfunc="sum", margins = True)
```

```python
pd.crosstab(dfsales["Region"], dfsales["Type"], 
            values = dfsales["Units"], aggfunc="sum")
```
